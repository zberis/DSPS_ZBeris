##Part 1:

```python
np.random.seed(6220)
Norm_0 = stats.norm.rvs(0, 1, 1000)
Norm_10 = stats.norm.rvs(10, 1, 1000)
tStudent_0 = stats.t.rvs(2, 0, size = 1000)
tStudent_10 = stats.t.rvs(2, 10, size = 1000)
```

```python
fig, ax = plt.subplots(figsize=(6, 6), dpi = 100)
plt.hist(Norm_0);
plt.title("Normal Distribution with mean = 0 and variance = 1")
plt.ylabel("Frequency")
plt.xlabel("Value")
plt.grid(True, linestyle="--", linewidth=0.5, color="gray", alpha=0.3)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
```

```python
fig, ax = plt.subplots(figsize=(6, 6), dpi = 100)
plt.hist(Norm_10);
plt.title("Normal Distribution with mean = 10 and variance = 1")
plt.ylabel("Frequency")
plt.xlabel("Value")
plt.grid(True, linestyle="--", linewidth=0.5, color="gray", alpha=0.3)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
```

```python
fig, ax = plt.subplots(figsize=(6, 6), dpi = 100)
plt.hist(tStudent_0, bins = 100, range = [-50,50]);
plt.title("tStudent Distribution with 2 degrees of freedom centered on 0")
plt.ylabel("Frequency")
plt.xlabel("Value")
plt.grid(True, linestyle="--", linewidth=0.5, color="gray", alpha=0.3)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
```

```python
fig, ax = plt.subplots(figsize=(6, 6), dpi = 100)
plt.hist(tStudent_10, bins = 100);
plt.title("tStudent Distribution with 2 degrees of freedom centered on 10")
plt.ylabel("Frequency")
plt.xlabel("Value")
plt.grid(True, linestyle="--", linewidth=0.5, color="gray", alpha=0.3)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
```

##Part 2:

```python
def posterior(prior, likelihood):
  normalized_prior = (prior - prior.min()) / (prior.max() - prior.min())
  normalized_likelihood = (likelihood - likelihood.min()) / (likelihood.max() - likelihood.min())
  this_posterior =  normalized_prior*normalized_likelihood

  plt.plot(normalized_prior, label = "Prior", color = "black", linestyle = "--")
  plt.plot(normalized_likelihood, label = "Likelihood", color = "black")
  plt.plot(this_posterior, label = "Posterior", color = "red")
  plt.legend()
  plt.xlabel("probability") #all plots need axis labels!!
  plt.ylabel("probability") #all plots need axis labels!!
```

```python
#function call
posterior(tStudent_0, tStudent_10)
```
