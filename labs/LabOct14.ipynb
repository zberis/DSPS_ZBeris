```python
import numpy as np
from matplotlib import pylab as plt
```

```python
np.random.seed(123)
X = (np.random.randn(100) * 3.5) + 12
```

```python
X.mean()
```

```python
plt.hist(X, bins = 10)
```

```python
X_minmax = X - X.min()
plt.hist(X_minmax, bins = 10)
```

```python
X_minmax = X_minmax / X_minmax.max()
plt.hist(X_minmax, bins = 10)
```

```python
X_standard = X - X.mean()
plt.hist(X_standard, bins = 10)
```

```python
X_standard.mean()
```
